```python
import pandas as pd
from csv import DictReader
from csv import DictWriter
import csv
counter = 0
allEpisodes = []

with open('seinfeld_scripts.csv', 'r') as read_obj:
    csv_dict_reader = DictReader(read_obj)
    
    for row in csv_dict_reader:
        stop = False
        script2 = ""
        script3 = ""
        script4 = ""
        script5 = ""
        script6 = ""
        counterEnds = 0
        cleanedScript = ""

        for word in row['text'].split(' '):
            if "<" not in word:
                script2 = script2 + ' ' + word  

        for word in script2.split(' '):
            if len(word) > 0 and "\n" not in word:
                script4 = script4 + " " + word
        Written = 0;
        for word in script4.split(' '):
            tempWord = ""
            for letter in word:
                if letter == "." or letter == "!" or letter == "?" or letter=="&" or letter==";":
                    tempWord += ""
                else:
                    tempWord += letter
            if Written == 1:
                cleanedScript = cleanedScript + " " + tempWord
            if tempWord == "Written":
                Written = 1;
        cleanedScript = cleanedScript.replace("Larry David amp ","",1)    
        #print(cleanedScript)
        allEpisodes.append(cleanedScript)
        #print("------------------------------------------------------------------")
        
                
        counter += 1
        print(cleanedScript)
        if counter == 2:
            break
                

```

```python
with open('some.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    for text in allEpisodes:
        writer.writerow([text])
```
